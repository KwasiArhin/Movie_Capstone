```
from google.colab import drive,files 
from tensorflow.keras.models import load_model 
from google.colab.patches import cv2_imshow
import cv2
import json

drive.mount('/content/drive')
```

```
def make_prediction(model,classes):
  uploaded =  files.upload()
  file_name = list(uploaded.keys())[0]
  image= cv2.imread(file_name)
  resize_image = cv2.resize(image,(150,150)) * 1.0/255 
  reshaped = resize_image.reshape(1,150,150,3)
  cv2_imshow(image)
  print(classes)
  return model.predict_classes(reshaped)
```

```
with open('/content/drive/My Drive/Movie Images/Models/myclasses.json','r') as f:
  classes = json.load(f)
```

```
my_model = load_model('/content/drive/My Drive/Movie Images/Models/mymodel.h5')
```

```
make_prediction(my_model,classes)
```
